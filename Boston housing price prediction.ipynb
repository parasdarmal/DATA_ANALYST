```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

from matplotlib.animation import FuncAnimation

from sklearn.datasets import load_boston
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

from IPython.display import HTML


```

```python
boston = load_boston()
print(boston.DESCR)
```

Data Preprocessing

```python
features = pd.DataFrame(boston.data, columns = boston.feature_names)
features
```

```python
target = pd.DataFrame(boston.target, columns = ['target'])
#concatenate features and target into one dataframe. axis = 1 make it concatenate columnwise.
df = pd.concat([features,target],axis=1)
df
```

DATA VISUALISATION

```python
#use round (decimal=2) to set the precision to 2 decimal places.
df.describe().round(decimals = 2)
```

```python
#Calculate correlation for every column on the data
corr = df.corr('pearson')

#Take absolute value of correlation between each feature and target.
corrs = [abs(corr[attr]['target']) for attr in list(features)]

#Make a list of pairs [(corr,feature)]
l = list(zip(corrs,list(features)))

#Sort list in reverse descending order
#with the correlation value as the key for sorting
l.sort(key = lambda x : x[0], reverse=True)

# unzip pairs to two list
#zip(*l) - takes a list that looks like [[a,b,c], [d,e,f], [g,h,i]]
# and return [[a,d,g], [b,e,h], [c,f,i]]
corrs,labels = list(zip((*l)))

#Plot correlation with respect to target variable as the bar garph
index = np.arange(len(labels))
plt.figure(figsize=(15,5))
plt.bar(index, corrs, width = 0.5)
plt.xlabel('Attributes')
plt.ylabel('Correlation with the target variable')
plt.xticks(index, labels)
plt.show()

```

NORMALIZE THE DATA

```python
X = df['LSTAT'].values
Y = df['target'].values
```

```python
#Before Normalization
print(Y[:5])
```

```python
x_scaler = MinMaxScaler()
X = x_scaler.fit_transform(X.reshape(-1,1))
X = X[:,-1]
y_scaler = MinMaxScaler()
Y = y_scaler.fit_transform(Y.reshape(-1,1))
Y = Y[:,-1]
```

```python
print(Y[:5])
```

LOSS

```python
def error(m,x,c,t):
    N = x.size
    e = sum(((m*x + c) - t)**2)
    return e * 1/(2 * N)
```

Splitting the Data

```python
#0.2 indicates 20% of data
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.2)
```

```python
def update(m, x, c, t, learning_rate):
    grad_m = sum(2 * ((m * x + c) - t) * x)
    grad_c = sum(2 * ((m * x + c) - t))
    m = m - grad_m * learning_rate
    c = c - grad_c * learning_rate
    return m,c
```

```python
def gradient_descent(init_m, init_c, x, t, learning_rate, iterations, error_threshold):
    m = init_m
    c = init_c
    error_values = list()
    mc_values = list()
    for i in range(iterations):
        e = error(m, x, c, t)
        if e < error_threshold :
            print('error less than the threshold. stopping gradient descent')
            break
        error_values.append(e)
        m,c = update(m, x, c, t, learning_rate)
        mc_values.append((m, c))
    return m, c, error_values, mc_values
    
```

Training phase

```python
%%time
init_m = 0.9
init_c = 0
learning_rate = 0.001
iterations = 250
error_threshold = 0.001

m, c, error_values, mc_values = gradient_descent(init_m, init_c, x_train, y_train, learning_rate, iterations, error_threshold)
```

ANIMATION

```python
# As no. of iterations increases changes in line are less noticable
#Inorder to reduce the processing timefor the animation, it is advisable to choose small interval
mc_values_anim = mc_values[0:250:5]
```

```python
fig, ax = plt.subplots()
ln, = plt.plot([], [], 'ro-', animated = True)

def init():
    plt.scatter(x_test,y_test,color = 'g')
    ax.set_xlim(0, 1.0)
    ax.set_ylim(0, 1.0)
    return ln,

def update_frame(frame):
    m , c = mc_values_anim[frame]
    x1, y1 = -0.5, m * -.5 + c
    x2, y2 = -1.5, m * -1.5 + c
    ln.set_data([x1,y1], [x2,y2])
    return ln,

anim = FuncAnimation(fig, update_frame, frames = range(len(mc_values_anim)), init_func = init, blit = True)

HTML(anim.to_html5_video())

```

Prediction

```python
# Calculate the predictions on the test set as a vectorized operation
predicted = (m * x_test) + c
```

```python
# Compute MSE for the predicted values on the testing set
mean_squared_error(y_test, predicted)
```
